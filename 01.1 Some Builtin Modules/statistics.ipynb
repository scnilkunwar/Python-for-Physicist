### Functions in the `statistics` Module

1. **mean(data)**
   - Returns the arithmetic mean (average) of the given data.

2. **median(data)**
   - Returns the median (middle value) of the given data.

3. **mode(data)**
   - Returns the mode (most common value) of the given data.

4. **multimode(data)**
   - Returns a list of the most common values in the data.

5. **stdev(data, xbar=None)**
   - Returns the standard deviation of the given data. An optional mean value can be provided.

6. **pvariance(data, mu=None)**
   - Returns the population variance of the given data. An optional population mean can be provided.

7. **variance(data, xbar=None)**
   - Returns the sample variance of the given data. An optional mean value can be provided.

8. **pstdev(data, mu=None)**
   - Returns the population standard deviation of the given data. An optional population mean can be provided.

9. **median_low(data)**
   - Returns the low median (the smaller of the two middle values) of the given data.

10. **median_high(data)**
    - Returns the high median (the larger of the two middle values) of the given data.

11. **harmonic_mean(data)**
    - Returns the harmonic mean of the given data.

12. **geometric_mean(data)**
    - Returns the geometric mean of the given data.

13. **quantiles(data, n=4, method='exclusive')**
    - Returns a list of quantiles for the given data, where `n` is the number of quantiles.

14. **standard_deviation(data, xbar=None)**
    - Returns the sample standard deviation of the given data. An optional mean value can be provided.

15. **lognormvariate(mu, sigma)**
    - Returns a random float from a log-normal distribution with mean `mu` and standard deviation `sigma`.

```python
import statistics

# Example data
data = [1, 2, 2, 3, 4, 5, 5, 5, 6]

# Example of mean()
mean_value = statistics.mean(data)
print("Mean:", mean_value)

# Example of median()
median_value = statistics.median(data)
print("Median:", median_value)

# Example of mode()
mode_value = statistics.mode(data)
print("Mode:", mode_value)

# Example of multimode()
multi_mode = statistics.multimode(data)
print("Multimode:", multi_mode)

# Example of stdev()
stdev_value = statistics.stdev(data)
print("Standard Deviation:", stdev_value)

# Example of pvariance()
pvariance_value = statistics.pvariance(data)
print("Population Variance:", pvariance_value)

# Example of variance()
variance_value = statistics.variance(data)
print("Sample Variance:", variance_value)

# Example of pstdev()
pstdev_value = statistics.pstdev(data)
print("Population Standard Deviation:", pstdev_value)

# Example of median_low()
median_low_value = statistics.median_low(data)
print("Low Median:", median_low_value)

# Example of median_high()
median_high_value = statistics.median_high(data)
print("High Median:", median_high_value)

# Example of harmonic_mean()
harmonic_mean_value = statistics.harmonic_mean(data)
print("Harmonic Mean:", harmonic_mean_value)

# Example of geometric_mean()
geometric_mean_value = statistics.geometric_mean(data)
print("Geometric Mean:", geometric_mean_value)

# Example of quantiles()
quantiles_value = statistics.quantiles(data, n=4)
print("Quantiles:", quantiles_value)
```
