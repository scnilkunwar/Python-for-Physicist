For example:

python
Copy code

result = arr + matrix
print(result)
# Output:
# [[2 4 6]
#  [5 7 9]]
Here, NumPy â€œbroadcastsâ€ the shape of arr across matrix by treating arr as a row vector, then adding it to each row of matrix.

17. Vectorization with NumPy

python
Copy code

28. Memory Management and Efficiency
a. View vs Copy in NumPy
When manipulating arrays, NumPy can either create a view or a copy of the original array. Understanding the difference is crucial for efficient memory management.

i. Views A view is a new array object that looks at the same data as the original array. Modifications to the view affect the original array.

python
Copy code
arr = np.array([1, 2, 3, 4])
view = arr[:2]
view[0] = 10
print(arr)  # Output: [10  2  3  4]
ii. Copies A copy is a new array object with its own data. Changes to the copy do not affect the original array.

python
Copy code
arr = np.array([1, 2, 3, 4])
copy = arr[:2].copy()
copy[0] = 10
print(arr)  # Output: [1 2 3 4]
b. Memory Layout: C vs F Order Revisited
Arrays in NumPy are stored in contiguous memory blocks in either row-major order (C-style) or column-major order (Fortran-style).

C-order: Row-major order, where rows are stored one after the other.
F-order: Column-major order, where columns are stored one after the other.
You can specify the memory order when creating or reshaping arrays:

python
Copy code
arr_c = np.array([[1, 2], [3, 4]], order='C')
arr_f = np.array([[1, 2], [3, 4]], order='F')
print(arr_c.flags)  # Output: C_CONTIGUOUS: True
print(arr_f.flags)  # Output: F_CONTIGUOUS: True
