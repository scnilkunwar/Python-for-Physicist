# SymPy `stats` Module Functions

## 1. **Random Variables**
- `sympy.stats.Bernoulli()`: Bernoulli random variable.
- `sympy.stats.Binomial()`: Binomial random variable.
- `sympy.stats.Poisson()`: Poisson random variable.
- `sympy.stats.Geometric()`: Geometric random variable.
- `sympy.stats.Hypergeometric()`: Hypergeometric random variable.
- `sympy.stats.NegativeBinomial()`: Negative Binomial random variable.
- `sympy.stats.Uniform()`: Uniform random variable (continuous).
- `sympy.stats.DiscreteUniform()`: Uniform random variable (discrete).
- `sympy.stats.Normal()`: Normal (Gaussian) random variable.
- `sympy.stats.Exponential()`: Exponential random variable.
- `sympy.stats.Gamma()`: Gamma random variable.
- `sympy.stats.Beta()`: Beta random variable.
- `sympy.stats.Pareto()`: Pareto random variable.
- `sympy.stats.Laplace()`: Laplace random variable.
- `sympy.stats.Rayleigh()`: Rayleigh random variable.
- `sympy.stats.Weibull()`: Weibull random variable.
- `sympy.stats.LogNormal()`: Log-normal random variable.
- `sympy.stats.Triangular()`: Triangular random variable.
- `sympy.stats.ChiSquared()`: Chi-squared random variable.
- `sympy.stats.StudentT()`: Student's t random variable.
- `sympy.stats.FDistribution()`: F-distribution random variable.
- `sympy.stats.VonMises()`: Von Mises random variable.

## 2. **Joint Distributions**
- `sympy.stats.JointRV()`: Define a joint random variable.
- `sympy.stats.JointDistribution()`: General joint distribution.
- `sympy.stats.MultivariateNormal()`: Multivariate normal random variable.
- `sympy.stats.Dirichlet()`: Dirichlet random variable.
- `sympy.stats.Multinomial()`: Multinomial random variable.

## 3. **Functions for Working with Random Variables**
- `sympy.stats.density()`: Probability density function (PDF) or probability mass function (PMF).
- `sympy.stats.cdf()`: Cumulative distribution function (CDF).
- `sympy.stats.sample()`: Generate a random sample from a distribution.
- `sympy.stats.where()`: Conditional distribution.
- `sympy.stats.pmf()`: Probability mass function for discrete random variables.
- `sympy.stats.pdf()`: Probability density function for continuous random variables.

## 4. **Probability and Expectation**
- `sympy.stats.Probability()`: Probability of an event.
- `sympy.stats.P()`: Alias for `Probability()`.
- `sympy.stats.Expectation()`: Expected value of a random variable.
- `sympy.stats.E()`: Alias for `Expectation()`.

## 5. **Variance and Moments**
- `sympy.stats.Variance()`: Variance of a random variable.
- `sympy.stats.var()`: Alias for `Variance()`.
- `sympy.stats.StandardDeviation()`: Standard deviation of a random variable.
- `sympy.stats.std()`: Alias for `StandardDeviation()`.
- `sympy.stats.Covariance()`: Covariance between two random variables.
- `sympy.stats.cov()`: Alias for `Covariance()`.
- `sympy.stats.Skewness()`: Skewness of a random variable.
- `sympy.stats.skewness()`: Alias for `Skewness()`.
- `sympy.stats.Kurtosis()`: Kurtosis of a random variable.
- `sympy.stats.kurtosis()`: Alias for `Kurtosis()`.

## 6. **Transformations**
- `sympy.stats.moment()`: nth moment of a random variable.
- `sympy.stats.cmoment()`: Central moment of a random variable.
- `sympy.stats.quantile()`: Quantile function.

## 7. **Conditional Probability and Expectation**
- `sympy.stats.Conditional()`: Define a conditional distribution.
- `sympy.stats.conditional()`: Compute a conditional probability or expectation.
- `sympy.stats.Piecewise()`: Piecewise-defined function for conditional probabilities.

## 8. **Utility Functions**
- `sympy.stats.sample_iter()`: Infinite iterator for random samples.
- `sympy.stats.density_for()`: Retrieve the density for a given variable.
- `sympy.stats.pspace()`: Probability space of a random expression.
- `sympy.stats.random_symbols()`: Retrieve random symbols in an expression.
