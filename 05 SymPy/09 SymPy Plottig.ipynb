## Plotting and Visualization

In this section, we will explore plotting and visualization capabilities in SymPy. We will cover basic plotting, advanced plotting techniques, and ways to customize plots. Each subsection includes examples and explanations.

### Basic Plotting with SymPy

SymPy provides an easy way to create simple plots using the `plot()` function.

```python
import sympy as sp

# Define a symbol
x = sp.symbols('x')

# Define a function
f = sp.sin(x)

# Plot the function
sp.plot(f, (x, -2*sp.pi, 2*sp.pi), title='Plot of f(x) = sin(x)', ylabel='f(x)')
```

### Advanced Plotting Techniques
SymPy allows you to create multiple plots, parametric plots, and more sophisticated visualizations.

```python
# Define another function
g = sp.cos(x)

# Create multiple plots
sp.plot(f, g, (x, -2*sp.pi, 2*sp.pi), legend=True, show=True)

# Parametric plot
t = sp.symbols('t')
parametric_plot = sp.plot_parametric(sp.cos(t), sp.sin(t), (t, 0, 2*sp.pi), title='Parametric Plot: Circle', ylabel='y', xlabel='x')
```

### Customizing Plots
You can customize various aspects of the plots, including titles, labels, colors, and more.

```python
# Create a customized plot
from matplotlib.pyplot import grid


p = sp.plot(f, (x, -2*sp.pi, 2*sp.pi), title='Customized Plot of f(x) = sin(x)', 
            ylabel='f(x)', xlabel='x', line_color='blue', show=True, legend=True)

# Customize the line style
p[0].line_color = 'green'
p[0].label = 'sin(x)'  # Label for the legend
```
