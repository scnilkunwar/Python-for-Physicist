# Matplotlib Customization Functions

## 1. Figure and Axes Customization
- **`plt.figure()`**  
  - Creates a new figure.
  - **Parameters**:
    - `figsize`: Tuple specifying the width and height of the figure in inches, e.g., `figsize=(10, 6)`.
    - `dpi`: Resolution of the figure in dots per inch, e.g., `dpi=100`.
    - `facecolor`: Background color of the figure, e.g., `facecolor='white'`.
    - `edgecolor`: Color of the figureâ€™s border, e.g., `edgecolor='black'`.
    - `num`: Figure number or label, e.g., `num=1`.

- **`fig.add_subplot()`**  
  - Adds a subplot to an existing figure.
  - **Parameters**:
    - `nrows`, `ncols`: Number of rows and columns for the subplot grid.
    - `index`: Position of the subplot in the grid.
    - `projection`: Type of plot, e.g., `projection='3d'` for 3D plotting.

- **`plt.subplots()`**  
  - Creates multiple subplots.
  - **Parameters**:
    - `nrows`, `ncols`: Number of rows and columns.
    - `figsize`, `dpi`: Size and resolution of the figure.
    - `sharex`, `sharey`: Share x-axis and y-axis among subplots.
    - `gridspec_kw`: Dictionary to customize grid layout, e.g., `gridspec_kw={'width_ratios': [2, 1]}`.

## 2. Plot Appearance
- **`plt.plot()`**  
  - Basic line plot with options for line style, color, markers, and more.
  - **Parameters**:
    - `x`, `y`: Data for x and y axes.
    - `color`: Color of the line, e.g., `'blue'`.
    - `linestyle`: Line style, e.g., `'--'` for dashed, `'-.'` for dash-dot.
    - `linewidth`: Width of the line in points.
    - `marker`: Shape of markers, e.g., `'o'` for circles, `'^'` for triangles.
    - `label`: Label for legend.

- **`plt.scatter()`**  
  - Creates a scatter plot.
  - **Parameters**:
    - `x`, `y`: Data for x and y axes.
    - `s`: Size of markers, e.g., `s=20`.
    - `c`: Color of markers, can be a single color or an array for colormap.
    - `alpha`: Transparency level, between 0 (transparent) and 1 (opaque).
    - `marker`: Shape of each point, e.g., `'x'`, `'o'`.
    - `label`: Label for legend.

- **`plt.bar()`**  
  - Creates a bar plot.
  - **Parameters**:
    - `x`, `height`: Positions of bars on x-axis and their heights.
    - `width`: Width of bars, default is `0.8`.
    - `color`: Color of the bars.
    - `edgecolor`: Color of bar edges.
    - `align`: Alignment of bars, `'center'` or `'edge'`.

## 3. Axis Customization
- **`ax.set_xlim()` / `ax.set_ylim()`**  
  - Sets the limits for x or y axis.
  - **Parameters**:
    - `left`, `right` / `bottom`, `top`: Left and right (or bottom and top) axis limits.

- **`ax.set_xticks()` / `ax.set_yticks()`**  
  - Specifies tick positions on the x or y axis.
  - **Parameters**:
    - `ticks`: List of positions where ticks should appear, e.g., `[0, 0.5, 1.0]`.

- **`plt.xticks()` / `plt.yticks()`**  
  - Sets custom tick labels and positions for x or y axis.
  - **Parameters**:
    - `ticks`: Positions for the ticks.
    - `labels`: Labels for each position.
    - `rotation`: Angle of label rotation, e.g., `rotation=45`.

- **`ax.grid()`**  
  - Adds grid lines to the plot.
  - **Parameters**:
    - `visible`: Set to `True` to show grid.
    - `color`: Grid color.
    - `linestyle`: Style of grid lines, e.g., `'-'`, `'--'`.
    - `linewidth`: Width of grid lines.

## 4. Adding Annotations and Text
- **`plt.text()`**  
  - Adds text to the plot at a specified location.
  - **Parameters**:
    - `x`, `y`: Coordinates of the text.
    - `s`: Text string.
    - `fontsize`: Size of text.
    - `color`: Color of text.

- **`plt.annotate()`**  
  - Adds annotations with optional arrows.
  - **Parameters**:
    - `text`: The annotation text.
    - `xy`: Point to annotate.
    - `xytext`: Position for text.
    - `arrowprops`: Dictionary of arrow properties, e.g., `{'arrowstyle': '->', 'color': 'red'}`.

- **`plt.legend()`**  
  - Adds a legend to the plot.
  - **Parameters**:
    - `loc`: Location, e.g., `'upper right'`.
    - `fontsize`: Size of legend text.
    - `frameon`: Draw a frame around legend (`True` or `False`).

## 5. Color and Style Customization
- **`plt.style.use()`**  
  - Applies a predefined style.
  - **Parameters**:
    - `style`: Style name, e.g., `'ggplot'`, `'seaborn'`, `'fivethirtyeight'`.

- **`plt.colorbar()`**  
  - Adds a color bar to the plot.
  - **Parameters**:
    - `mappable`: The image or data source for color scaling.
    - `orientation`: `'vertical'` or `'horizontal'`.
    - `label`: Label for the colorbar.

## 6. Line and Marker Customization
- **`plt.Line2D()`**  
  - Directly customize lines.
  - **Parameters**:
    - `xdata`, `ydata`: Data points.
    - `color`, `linestyle`, `linewidth`: Line properties.
    - `marker`, `markersize`, `markeredgecolor`: Marker properties.

## 7. Fonts and Text Customization
- **`plt.rcParams.update()`**  
  - Updates default settings.
  - **Parameters**:
    - Dictionary with font family, size, and other options.
    - Example: `plt.rcParams.update({'font.size': 14, 'font.family': 'serif'})`.

## 8. 3D Plot Customization
- **`ax.plot_surface()`**  
  - Creates a 3D surface plot.
  - **Parameters**:
    - `X`, `Y`, `Z`: Data arrays for x, y, and z.
    - `cmap`: Colormap for surface, e.g., `'viridis'`.
    - `edgecolor`: Color of edges.

- **`ax.view_init()`**  
  - Adjusts view angle for 3D plots.
  - **Parameters**:
    - `elev`: Elevation angle.
    - `azim`: Azimuth angle.

## 9. Saving Figures
- **`plt.savefig()`**  
  - Saves the figure to a file.
  - **Parameters**:
    - `fname`: File path or name.
    - `dpi`: Resolution in dots per inch.
    - `format`: File format, e.g., `'png'`, `'pdf'`.
    - `transparent`: If `True`, makes background transparent.
    - `bbox_inches`: Extends bounding box, e.g., `'tight'` to crop.

## 10. Using Matplotlib Styles
- **`plt.style.use()`**  
  - Applies a predefined Matplotlib style.
  - **Parameters**:
    - `style`: String name of the style, e.g., `'seaborn-darkgrid'`, `'classic'`, or `'ggplot'`.


```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Figure and Axes Customization
# Create a figure with custom size and DPI
fig = plt.figure(figsize=(8, 6), dpi=120, facecolor='lightgray')
ax = fig.add_subplot(1, 1, 1)  # Adds a subplot

# Using plt.subplots to create a grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), dpi=100, sharex=True, sharey=True)
plt.show()


# 2. Plot Appearance
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line plot with custom line color, style, and markers
plt.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o', label='Sine wave')
plt.legend()
plt.show()

# Scatter plot with custom color, size, and transparency
plt.scatter(x, y, c='red', s=50, alpha=0.7, label='Scatter points')
plt.legend()
plt.show()

# Bar plot with custom width and edge color
x = np.arange(5)
heights = [5, 7, 3, 8, 6]
plt.bar(x, heights, width=0.6, color='skyblue', edgecolor='black')
plt.show()


# 3. Axis Customization
fig, ax = plt.subplots()

# Set axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

# Set custom tick positions and labels
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, 0, 1])
plt.show()

# Add grid with specific color and line style
ax.grid(visible=True, color='gray', linestyle='--', linewidth=0.5)
plt.show()


# 4. Adding Annotations and Text
plt.plot(x, y)

# Add text at specific location
plt.text(5, 0, "Center Point", fontsize=12, color='blue')

# Add annotation with arrow
plt.annotate("Peak", xy=(1.5 * np.pi, 1), xytext=(4, 1.2),
             arrowprops=dict(arrowstyle='->', color='green'))

plt.legend(['Sine wave'])
plt.show()


# 5. Color and Style Customization
# Set a style
plt.style.use('seaborn-darkgrid')
plt.plot(x, y, label='Styled Plot')
plt.legend()
plt.show()

# Adding a colorbar
data = np.random.rand(10, 10)
plt.imshow(data, cmap='viridis')
plt.colorbar(label='Intensity')
plt.show()


# 6. Line and Marker Customization
# Create a line with specific properties
line = plt.Line2D(x, y, color='orange', linestyle='-', linewidth=2, marker='x', markersize=8)
fig, ax = plt.subplots()
ax.add_line(line)
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)
plt.show()


# 7. Fonts and Text Customization
# Set default font size and family
plt.rcParams.update({'font.size': 14, 'font.family': 'serif'})
plt.plot(x, y, label='Sine wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()


# 8. 3D Plot Customization
from mpl_toolkits.mplot3d import Axes3D

# Creating 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')

# Adjust view angle
ax.view_init(elev=45, azim=30)
plt.show()


# 9. Saving Figures
# Save figure with specific DPI and format
plt.plot(x, y)
plt.savefig('sine_wave.png', dpi=150, format='png', transparent=True, bbox_inches='tight')


# 10. Using Matplotlib Styles
# Set a predefined style
plt.style.use('ggplot')
plt.plot(x, y, label='Styled with ggplot')
plt.legend()
plt.show()
```
