# Matplotlib Plotting Functions with Parameters

## 1. Basic Line and Scatter Plots
- `plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o')`
  - **Description**: Creates a line plot or scatter plot.
  - **Parameters**:
    - **x**: X-axis data.
    - **y**: Y-axis data.
    - **color**: Color of the line.
    - **linestyle**: Style of the line (e.g., `'-'`, `'--'`).
    - **linewidth**: Width of the line.
    - **marker**: Marker style (e.g., `'o'`, `'^'`).

```python
# 1. Basic Line and Scatter Plots

# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o')
plt.title("Line Plot")
plt.show()
```

- `scatter(x, y, color='red', s=20, marker='*', alpha=0.5)`
  - **Description**: Creates a scatter plot of points.
  - **Parameters**:
    - **x**: X-axis data.
    - **y**: Y-axis data.
    - **color**: Marker color.
    - **s**: Marker size.
    - **marker**: Marker style.
    - **alpha**: Transparency level.

```python
# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.scatter(x, y, color='red', s=100, marker='*', alpha=0.5)
plt.title("Scatter Plot")
plt.show()
```

- `step(x, y, where='mid', color='purple')`
  - **Description**: Creates a step plot.
  - **Parameters**:
    - **where**: Step position (`'pre'`, `'mid'`, `'post'`).

```python
# Step Plot
plt.figure()
plt.step(np.sort(x), y, where='mid', color='purple')
plt.title("Step Plot")
plt.show()
```

- `stem(x, y, linefmt='g-', markerfmt='bo', basefmt='r-')`
  - **Description**: Creates a stem plot with vertical lines.
  - **Parameters**:
    - **linefmt**: Format for the lines.
    - **markerfmt**: Format for the markers.
    - **basefmt**: Format for the baseline.

```python
# Stem Plot
plt.figure()
plt.stem(x[:10], y[:10], linefmt='g-', markerfmt='bo', basefmt='r-')
plt.title("Stem Plot")
plt.show()
```

- `errorbar(x, y, yerr=0.2, xerr=0.1, fmt='o', ecolor='black', capsize=5)`
  - **Description**: Creates a plot with error bars.
  - **Parameters**:
    - **yerr**: Y error values.
    - **xerr**: X error values.
    - **fmt**: Format for the plot (e.g., `'o'` for circles).
    - **ecolor**: Color of the error bars.
    - **capsize**: Size of the error bar caps.

```python
# Error Bar Plot
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.1 * np.abs(np.random.randn(10))
plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='black', capsize=5)
plt.title("Error Bar Plot")
plt.show()
```

## 2. Histograms and Density Plots
- `hist(data, bins=10, color='blue', edgecolor='black', alpha=0.7)`
  - **Description**: Creates a histogram.
  - **Parameters**:
    - **data**: Data values.
    - **bins**: Number of bins.
    - **color**: Bar color.
    - **edgecolor**: Color of bin edges.
    - **alpha**: Transparency level.

```python
# 2. Histograms and Density Plots

# Histogram
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title("Histogram")
plt.show()
```

- `bar(x, height, width=0.8, color='green', edgecolor='black')`
  - **Description**: Creates a vertical bar plot.
  - **Parameters**:
    - **x**: X-axis data.
    - **height**: Bar height.
    - **width**: Width of each bar.
    - **color**: Bar color.
    - **edgecolor**: Color of bar edges.

```python
# Bar Plot
x = ['A', 'B', 'C', 'D']
height = [3, 7, 5, 4]
plt.figure()
plt.bar(x, height, width=0.4, color='green', edgecolor='black')
plt.title("Bar Plot")
plt.show()
```

- `barh(y, width, height=0.8, color='purple')`
  - **Description**: Creates a horizontal bar plot.
  - **Parameters**:
    - **y**: Y-axis data.
    - **width**: Width of each bar.
    - **height**: Height of each bar.

```python
# Horizontal Bar Plot
plt.figure()
plt.barh(x, height, color='purple')
plt.title("Horizontal Bar Plot")
plt.show()
```

- `boxplot(data, vert=True, patch_artist=True, notch=False)`
  - **Description**: Creates a box plot to show distributions.
  - **Parameters**:
    - **data**: Data values.
    - **vert**: Vertical orientation (True/False).
    - **patch_artist**: Fill boxes with color (True/False).
    - **notch**: Draw a notch (True/False).

```python
# Box Plot
data = [np.random.normal(size=100) for _ in range(4)]
plt.figure()
plt.boxplot(data, vert=True, patch_artist=True, notch=False)
plt.title("Box Plot")
plt.show()
```

- `violinplot(data, showmeans=True, showmedians=True)`
  - **Description**: Creates a violin plot to show distribution shape.
  - **Parameters**:
    - **data**: Data values.
    - **showmeans**: Show mean values (True/False).
    - **showmedians**: Show median values (True/False).

```python
# Violin Plot
plt.figure()
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title("Violin Plot")
plt.show()
```

## 3. Pie and Polar Plots
- `pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)`
  - **Description**: Creates a pie chart.
  - **Parameters**:
    - **sizes**: Data values for slices.
    - **labels**: Labels for each slice.
    - **autopct**: String format for percentages.
    - **startangle**: Starting angle for the first slice.

```python
# 3. Pie and Polar Plots

# Pie Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```

- `polar(theta, r, color='blue')`
  - **Description**: Creates a polar plot.
  - **Parameters**:
    - **theta**: Angular coordinate.
    - **r**: Radial coordinate.

```python
# Polar Plot
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta))
plt.figure()
plt.polar(theta, r, color='blue')
plt.title("Polar Plot")
plt.show()
```

## 4. Statistical and Relationship Plots
- `hexbin(x, y, gridsize=30, cmap='Blues')`
  - **Description**: Creates a hexagonal binning plot.
  - **Parameters**:
    - **x**: X-axis data.
    - **y**: Y-axis data.
    - **gridsize**: Size of the hexagonal grid.
    - **cmap**: Colormap for the hexagons.

```python
# 4. Statistical and Relationship Plots

# Hexbin Plot
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.figure()
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.title("Hexbin Plot")
plt.colorbar()
plt.show()
```

- `contour(X, Y, Z, levels=10, cmap='viridis')`
  - **Description**: Creates a contour plot.
  - **Parameters**:
    - **X**: X-axis grid values.
    - **Y**: Y-axis grid values.
    - **Z**: Function values at each (X, Y) point.
    - **levels**: Number of contour levels.
    - **cmap**: Colormap for the contours.

```python

# Contour Plot
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.exp(-X**2 - Y**2)
plt.figure()
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.title("Contour Plot")
plt.show()
```

- `contourf(X, Y, Z, levels=10, cmap='viridis')`
  - **Description**: Creates a filled contour plot.
  - **Parameters**: Same as `contour()`, but fills the areas between contours.

```python

# Filled Contour Plot
plt.figure()
plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

- `imshow(data, cmap='gray', interpolation='nearest')`
  - **Description**: Displays an image or 2D data as a grid.
  - **Parameters**:
    - **data**: 2D array of data values.
    - **cmap**: Colormap to apply.
    - **interpolation**: Interpolation method (e.g., `'nearest'`, `'bilinear'`).

```python

# Image Plot
data = np.random.rand(10, 10)
plt.figure()
plt.imshow(data, cmap='gray', interpolation='nearest')
plt.title("Image Plot")
plt.colorbar()
plt.show()
```

## 5. 3D Plots (with `mpl_toolkits.mplot3d`)
- `plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')`
  - **Description**: Creates a 3D surface plot.
  - **Parameters**:
    - **X**: X-axis grid values.
    - **Y**: Y-axis grid values.
    - **Z**: Z-axis height values.
    - **cmap**: Colormap for the surface.
    - **edgecolor**: Color of surface edges.

```python

# 5. 3D Plots (with mpl_toolkits.mplot3d)
from mpl_toolkits.mplot3d import Axes3D

# 3D Surface Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("3D Surface Plot")
plt.show()
```


- `scatter3D(x, y, z, color='red', s=20)`
  - **Description**: Creates a 3D scatter plot.
  - **Parameters**:
    - **x**: X-axis data.
    - **y**: Y-axis data.
    - **z**: Z-axis data.
    - **color**: Marker color.
    - **s**: Marker size.

```python

# 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
ax.scatter(x, y, z, color='red', s=20)
ax.set_title("3D Scatter Plot")
plt.show()
```

- `plot_wireframe(X, Y, Z, color='blue', linewidth=0.5)`
  - **Description**: Creates a 3D wireframe plot.
  - **Parameters**:
    - **X**: X-axis grid values.
    - **Y**: Y-axis grid values.
    - **Z**: Z-axis height values.
    - **color**: Color of the lines.
    - **linewidth**: Width of the lines.

```python

# 3D Wireframe Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='blue', linewidth=0.5)
ax.set_title("3D Wireframe Plot")
plt.show()
```

## 6. Advanced and Specialty Plots
- `quiver(X, Y, U, V, color='blue')`
  - **Description**: Creates a 2D vector field plot.
  - **Parameters**:
    - **X**: X-axis grid values.
    - **Y**: Y-axis grid values.
    - **U**: X components of the vectors.
    - **V**: Y components of the vectors.
    - **color**: Color of the arrows.

```python
# 6. Advanced and Specialty Plots

# Quiver Plot
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
plt.figure()
plt.quiver(X, Y, U, V, color='blue')
plt.title("Quiver Plot")
plt.show()
```

- `streamplot(X, Y, U, V, color='blue')`
  - **Description**: Creates streamlines of a vector field.
  - **Parameters**:
    - **X**: X-axis grid values.
    - **Y**: Y-axis grid values.
    - **U**: X components of the vectors.
    - **V**: Y components of the vectors.

```python

# Streamline Plot
plt.figure()
plt.streamplot(X, Y, U, V, color='blue')
plt.title("Streamline Plot")
plt.show()
```

- `fill_between(x, y1, y2, color='orange', alpha=0.5)`
  - **Description**: Fills the area between two curves.
  - **Parameters**:
    - **x**: X-axis data.
    - **y1**: Data for the upper curve.
    - **y2**: Data for the lower curve.
    - **color**: Fill color.
    - **alpha**: Transparency level.

```python
# Fill Between Plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5
plt.figure()
plt.fill_between(x, y1, y2, color='orange', alpha=0.5)
plt.title("Fill Between Plot")
plt.show()
```

- `eventplot(data, orientation='horizontal', color='black')`
  - **Description**: Creates a plot for event data.
  - **Parameters**:
    - **data**: List of lists, where each inner list represents events for one group.
    - **orientation**: Orientation of the events (`'horizontal'` or `'vertical'`).
    - **color**: Color of the events.

```python
# Event Plot
data = [np.random.rand(5) for _ in range(3)]
plt.figure()
plt.eventplot(data, orientation='horizontal', color='black')
plt.title("Event Plot")
plt.show()
```
