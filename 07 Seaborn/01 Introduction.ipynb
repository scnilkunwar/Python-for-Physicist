# Seaborn Tutorial

## 1. Introduction to Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn is particularly well-suited for visualizing complex datasets and exploring relationships among variables.

## 2. Installation

To install Seaborn, you can use pip, which is the standard package manager for Python:

```bash
pip install seaborn
```
After installation, import Seaborn and Matplotlib in your Python script or Jupyter Notebook:

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

## 3. Setting Up Seaborn
### 3.1 Setting the Aesthetic
You can change the visual appearance of your plots using the sns.set() function. Seaborn has several built-in styles:

````python
sns.set(style="whitegrid")  # Other options: "white", "dark", "darkgrid", "ticks"
````

### 3.2 Color Palettes
Seaborn supports a variety of color palettes to enhance the aesthetics of your visualizations. You can set a color palette using:

````python
sns.set_palette("pastel")  # Other options: "deep", "muted", "bright", "dark", "colorblind"
````


You can also create custom color palettes:

```python
custom_palette = sns.color_palette("husl", 8)  # Create a custom palette with 8 colors
sns.set_palette(custom_palette)

### 4. Loading Example Datasets
Seaborn comes with several built-in datasets that you can easily load for practice:

````python
tips = sns.load_dataset("tips")  # Load the 'tips' dataset
iris = sns.load_dataset("iris")   # Load the 'iris' dataset
print(tips.head())  # Display the first few rows of the tips dataset
````

## 5. Plot Types in Seaborn
### 5.1 Distribution Plots
Distribution plots help visualize the distribution of a single continuous variable.

### 5.1.1 Histogram with KDE

````python
plt.figure(dpi = 400)
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title("Histogram of Total Bill with KDE")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
````

### 5.1.2 Kernel Density Estimate (KDE) Plot

````python
sns.kdeplot(tips['total_bill'], fill=True, color='purple')
plt.title("KDE Plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
````

````python
### 5.1.3 Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
````

### 5.2 Categorical Plots
Categorical plots are used to visualize the relationship between categorical and continuous variables.


````python
## 5.2.1 Bar Plot

sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill by Day")
plt.show()
````

````python
# 5.2.2 Count Plot

sns.countplot(x="day", data=tips)
plt.title("Count of Tips by Day")
plt.show()
````

````python
## 5.2.3 Box Plot

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
````

### 5.3 Relationship Plots
Relationship plots help visualize the relationship between two continuous variables.


````python
plt.figure(dpi = 400)
## 5.3.1 Scatter Plot

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()

## 5.3.2 Line Plot

sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line Plot of Total Bill by Party Size")
plt.show()

## 5.3.3 Regression Plot

sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot of Total Bill vs. Tip")
plt.show()
````

## 5.4 Matrix Plots
Matrix plots are useful for visualizing correlations and relationships in a grid format.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the tips dataset
tips = sns.load_dataset("tips")

# Select only the numeric columns for correlation
numeric_tips = tips.select_dtypes(include='number')

# 5.4.1 Heatmap
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap of Tips Dataset Correlation")
plt.show()

# 5.4.2 Cluster Map
clustermap = sns.clustermap(corr, annot=True, cmap='coolwarm')
clustermap.fig.suptitle("Clustermap of Tips Dataset Correlation", y=1.05)
plt.show()
````

## 5.5 Pairwise Plots
Pairwise plots show the pairwise relationships in a dataset.

````python
# 5.5.1 Pair Plot

sns.pairplot(tips, hue='sex')
plt.title("Pair Plot of Tips Dataset")
plt.show()

# 5.5.2 Pair Grid

g = sns.PairGrid(tips, hue='sex')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.title("Pair Grid of Tips Dataset")
plt.show()
````

## 8. Advanced Visualization Techniques
### 8.1 Combining Seaborn with Matplotlib
You can create complex visualizations by combining Seaborn and Matplotlib:



````python
sns.scatterplot(x="total_bill", y="tip", data=tips, color='blue')
plt.title("Scatter Plot with Matplotlib Customization")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid(True)
plt.show()
````

### 8.2 Multi-Plot Grids with sns.FacetGrid()
Facet grids allow you to create a grid of plots based on categorical variables:


````python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
````

### 8.3 Customizing Color Palettes
You can create a custom color palette using sns.color_palette():


````python
custom_palette = sns.color_palette("husl", 8)
sns.set_palette(custom_palette)
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot with Custom Color Palette")
plt.show()
````

### 8.4 Statistical Functions
Seaborn integrates statistical functions directly into plots:

````python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot with Statistical Summary")
plt.show()
````
