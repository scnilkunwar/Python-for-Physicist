# scipy.stats Module Overview

The `scipy.stats` module is a powerful library for statistical functions and probability distributions. It includes a wide range of functionalities, including statistical tests, descriptive statistics, and various probability distributions.

## 1. Probability Distributions

### Continuous Distributions
- **`scipy.stats.norm`**: Normal distribution.
  - **`pdf(x, loc=0, scale=1)`**: Probability density function.
  - **`cdf(x, loc=0, scale=1)`**: Cumulative distribution function.
  - **`rvs(size, loc=0, scale=1)`**: Generate random variates.

- **`scipy.stats.uniform`**: Uniform distribution.
- **`scipy.stats.expon`**: Exponential distribution.
- **`scipy.stats.gamma`**: Gamma distribution.
- **`scipy.stats.beta`**: Beta distribution.
- **`scipy.stats.chi2`**: Chi-squared distribution.
- **`scipy.stats.t`**: Student's t-distribution.
- **`scipy.stats.lognorm`**: Log-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon, gamma, beta, chi2, t, lognorm

# Normal Distribution
mean = 0
std_dev = 1
size = 1000

# Generate random variates for Normal distribution
samples_norm = norm.rvs(loc=mean, scale=std_dev, size=size)

# Parameters for Uniform distribution
low = 0
high = 1

# Generate random variates for Uniform distribution
samples_uniform = uniform.rvs(loc=low, scale=high - low, size=size)

# Parameters for Exponential distribution
scale_expon = 1

# Generate random variates for Exponential distribution
samples_expon = expon.rvs(scale=scale_expon, size=size)

# Parameters for Gamma distribution
shape_gamma = 2
scale_gamma = 1

# Generate random variates for Gamma distribution
samples_gamma = gamma.rvs(a=shape_gamma, scale=scale_gamma, size=size)

# Parameters for Beta distribution
a_beta = 2
b_beta = 5

# Generate random variates for Beta distribution
samples_beta = beta.rvs(a=a_beta, b=b_beta, size=size)

# Parameters for Chi-squared distribution
df_chi2 = 2

# Generate random variates for Chi-squared distribution
samples_chi2 = chi2.rvs(df=df_chi2, size=size)

# Parameters for Student's t-distribution
df_t = 2

# Generate random variates for Student's t-distribution
samples_t = t.rvs(df=df_t, size=size)

# Parameters for Log-normal distribution
shape_lognorm = 0.5

# Generate random variates for Log-normal distribution
samples_lognorm = lognorm.rvs(shape_lognorm, size=size)

# Plotting all distributions in a 3x3 grid
fig, axs = plt.subplots(3, 3, figsize=(15, 12))

# Normal Distribution Plot
axs[0, 0].hist(samples_norm, bins=30, density=True, alpha=0.6, color='g', label='Random Samples')
x_norm = np.linspace(-5, 5, 100)
pdf_norm = norm.pdf(x_norm, loc=mean, scale=std_dev)
axs[0, 0].plot(x_norm, pdf_norm, 'r', label='PDF')
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('Probability')
axs[0, 0].legend()

# Uniform Distribution Plot
axs[0, 1].hist(samples_uniform, bins=30, density=True, alpha=0.6, color='b', label='Random Samples')
x_uniform = np.linspace(0, 1, 100)
pdf_uniform = uniform.pdf(x_uniform, loc=low, scale=high - low)
axs[0, 1].plot(x_uniform, pdf_uniform, 'r', label='PDF')
axs[0, 1].set_title('Uniform Distribution')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('Probability')
axs[0, 1].legend()

# Exponential Distribution Plot
axs[0, 2].hist(samples_expon, bins=30, density=True, alpha=0.6, color='y', label='Random Samples')
x_expon = np.linspace(0, 10, 100)
pdf_expon = expon.pdf(x_expon, scale=scale_expon)
axs[0, 2].plot(x_expon, pdf_expon, 'r', label='PDF')
axs[0, 2].set_title('Exponential Distribution')
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('Probability')
axs[0, 2].legend()

# Gamma Distribution Plot
axs[1, 0].hist(samples_gamma, bins=30, density=True, alpha=0.6, color='c', label='Random Samples')
x_gamma = np.linspace(0, 10, 100)
pdf_gamma = gamma.pdf(x_gamma, a=shape_gamma, scale=scale_gamma)
axs[1, 0].plot(x_gamma, pdf_gamma, 'r', label='PDF')
axs[1, 0].set_title('Gamma Distribution')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('Probability')
axs[1, 0].legend()

# Beta Distribution Plot
axs[1, 1].hist(samples_beta, bins=30, density=True, alpha=0.6, color='m', label='Random Samples')
x_beta = np.linspace(0, 1, 100)
pdf_beta = beta.pdf(x_beta, a=a_beta, b=b_beta)
axs[1, 1].plot(x_beta, pdf_beta, 'r', label='PDF')
axs[1, 1].set_title('Beta Distribution')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('Probability')
axs[1, 1].legend()

# Chi-squared Distribution Plot
axs[1, 2].hist(samples_chi2, bins=30, density=True, alpha=0.6, color='g', label='Random Samples')
x_chi2 = np.linspace(0, 10, 100)
pdf_chi2 = chi2.pdf(x_chi2, df=df_chi2)
axs[1, 2].plot(x_chi2, pdf_chi2, 'r', label='PDF')
axs[1, 2].set_title('Chi-squared Distribution')
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('Probability')
axs[1, 2].legend()

# Student's t-Distribution Plot
axs[2, 0].hist(samples_t, bins=30, density=True, alpha=0.6, color='b', label='Random Samples')
x_t = np.linspace(-5, 5, 100)
pdf_t = t.pdf(x_t, df=df_t)
axs[2, 0].plot(x_t, pdf_t, 'r', label='PDF')
axs[2, 0].set_title('Student\'s t Distribution')
axs[2, 0].set_xlabel('x')
axs[2, 0].set_ylabel('Probability')
axs[2, 0].legend()

# Log-normal Distribution Plot
axs[2, 1].hist(samples_lognorm, bins=30, density=True, alpha=0.6, color='y', label='Random Samples')
x_lognorm = np.linspace(0, 10, 100)
pdf_lognorm = lognorm.pdf(x_lognorm, shape_lognorm)
axs[2, 1].plot(x_lognorm, pdf_lognorm, 'r', label='PDF')
axs[2, 1].set_title('Log-normal Distribution')
axs[2, 1].set_xlabel('x')
axs[2, 1].set_ylabel('Probability')
axs[2, 1].legend()

# Hide the empty subplot in the last position
axs[2, 2].axis('off')

# Adjust layout to ensure no overlap
plt.tight_layout()
plt.show()
```

### Discrete Distributions
- **`scipy.stats.binom`**: Binomial distribution.
- **`scipy.stats.poisson`**: Poisson distribution.
- **`scipy.stats.geom`**: Geometric distribution.
- **`scipy.stats.hypergeom`**: Hypergeometric distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, geom, hypergeom

# Binomial Distribution
# Parameters
n_binom = 10  # number of trials
p_binom = 0.5  # probability of success in each trial
size_binom = 1000  # number of random samples

# Generate random samples from binomial distribution
random_samples_binom = binom.rvs(n_binom, p_binom, size=size_binom)

# Probability mass function (PMF) for binomial distribution
k_binom = np.arange(0, n_binom + 1)  # number of successes
pmf_values_binom = binom.pmf(k_binom, n_binom, p_binom)

# Poisson Distribution
# Parameters
mu_poisson = 3  # average rate of events (mean)
size_poisson = 1000  # number of random samples

# Generate random samples from Poisson distribution
random_samples_poisson = poisson.rvs(mu_poisson, size=size_poisson)

# Probability mass function (PMF) for Poisson distribution
k_poisson = np.arange(0, max(random_samples_poisson) + 1)
pmf_values_poisson = poisson.pmf(k_poisson, mu_poisson)

# Geometric Distribution
# Parameters
p_geom = 0.3  # probability of success in each trial
size_geom = 1000  # number of random samples

# Generate random samples from geometric distribution
random_samples_geom = geom.rvs(p_geom, size=size_geom)

# Probability mass function (PMF) for geometric distribution
k_geom = np.arange(0, max(random_samples_geom) + 1)
pmf_values_geom = geom.pmf(k_geom, p_geom)

# Hypergeometric Distribution
# Parameters
M = 20  # total number of objects
n = 10  # number of successes in population
N = 5  # number of draws
size_hypergeom = 1000  # number of random samples

# Generate random samples from hypergeometric distribution
random_samples_hypergeom = hypergeom.rvs(M, n, N, size=size_hypergeom)

# Probability mass function (PMF) for hypergeometric distribution
k_hypergeom = np.arange(0, min(n, N) + 1)
pmf_values_hypergeom = hypergeom.pmf(k_hypergeom, M, n, N)

# Plotting the distributions in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Binomial Distribution Plot
axs[0, 0].hist(random_samples_binom, bins=10, density=True, alpha=0.6, color='g', label='Random Samples')
axs[0, 0].stem(k_binom, pmf_values_binom, 'r', label='PMF', basefmt=" ")
axs[0, 0].set_title('Binomial Distribution')
axs[0, 0].set_xlabel('Number of successes')
axs[0, 0].set_ylabel('Probability')
axs[0, 0].legend()

# Poisson Distribution Plot
axs[0, 1].hist(random_samples_poisson, bins=10, density=True, alpha=0.6, color='b', label='Random Samples')
axs[0, 1].stem(k_poisson, pmf_values_poisson, 'r', label='PMF', basefmt=" ")
axs[0, 1].set_title('Poisson Distribution')
axs[0, 1].set_xlabel('Number of events')
axs[0, 1].set_ylabel('Probability')
axs[0, 1].legend()

# Geometric Distribution Plot
axs[1, 0].hist(random_samples_geom, bins=10, density=True, alpha=0.6, color='y', label='Random Samples')
axs[1, 0].stem(k_geom, pmf_values_geom, 'r', label='PMF', basefmt=" ")
axs[1, 0].set_title('Geometric Distribution')
axs[1, 0].set_xlabel('Number of trials until success')
axs[1, 0].set_ylabel('Probability')
axs[1, 0].legend()

# Hypergeometric Distribution Plot
axs[1, 1].hist(random_samples_hypergeom, bins=10, density=True, alpha=0.6, color='m', label='Random Samples')
axs[1, 1].stem(k_hypergeom, pmf_values_hypergeom, 'r', label='PMF', basefmt=" ")
axs[1, 1].set_title('Hypergeometric Distribution')
axs[1, 1].set_xlabel('Number of successes in draws')
axs[1, 1].set_ylabel('Probability')
axs[1, 1].legend()

# Adjust layout to ensure no overlap
plt.tight_layout()
plt.show()
```

## 2. Random Variables
You can generate random samples from distributions and evaluate their properties:
- **`rvs(size, loc=0, scale=1)`**: Generate random variates from the distribution.
- **`pdf(x, loc=0, scale=1)`**: Probability density function.
- **`cdf(x, loc=0, scale=1)`**: Cumulative distribution function.
- **`ppf(q, loc=0, scale=1)`**: Percent point function (inverse of CDF).
- **`mean(loc=0, scale=1)`**: Mean of the distribution.
- **`std(loc=0, scale=1)`**: Standard deviation of the distribution.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

# Generate random samples
# 'rvs()' function generates 1000 random variates from a normal distribution with mean=0 and std=1
random_samples = norm.rvs(loc=0, scale=1, size=1000)

# Probability Density Function (PDF)
# 'pdf()' calculates the value of the probability density function at each point in the array 'x'
# 'x' is a set of points for which we want to evaluate the PDF
x = np.linspace(-5, 5, 100)
pdf_values = norm.pdf(x, loc=0, scale=1)

# Cumulative Distribution Function (CDF)
# 'cdf()' calculates the cumulative probability for each value in 'x'
cdf_values = norm.cdf(x, loc=0, scale=1)

# Percent Point Function (PPF)
# 'ppf()' returns the value of the distribution corresponding to a given cumulative probability (q)
q = np.linspace(0, 1, 100)
ppf_values = norm.ppf(q, loc=0, scale=1)

# Mean and Standard Deviation
# The 'mean()' function returns the mean of the normal distribution, which is 0 in this case
# The 'std()' function returns the standard deviation of the distribution, which is 1 in this case
mean_value = norm.mean(loc=0, scale=1)
std_dev = norm.std(loc=0, scale=1)

# Plotting all distributions in a 2x2 grid layout

# Plot 1: PDF of the normal distribution
plt.subplot(2, 2, 1)
plt.plot(x, pdf_values, label='Normal PDF')
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Plot 2: CDF of the normal distribution
plt.subplot(2, 2, 2)
plt.plot(x, cdf_values, label='Normal CDF', color='orange')
plt.title('Normal Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

# Plot 3: PPF of the normal distribution
plt.subplot(2, 2, 3)
plt.plot(q, ppf_values, label='Normal PPF', color='purple')
plt.title('Normal Distribution PPF')
plt.xlabel('Cumulative Probability (q)')
plt.ylabel('Value')

# Plot 4: Histogram of randomly generated samples from the normal distribution
plt.subplot(2, 2, 4)
plt.hist(random_samples, bins=30, density=True, alpha=0.6, color='g')
plt.title('Random Samples from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Adjust layout to prevent overlap and show all plots
plt.tight_layout()
plt.show()

# Output Mean and Standard Deviation
# Print the mean and standard deviation of the distribution
print(f'Mean: {mean_value}')
print(f'Standard Deviation: {std_dev}')
```

## 3. Descriptive Statistics
Functions to compute various descriptive statistics:
- **`scipy.stats.describe(a)`**: Computes several descriptive statistics.
- **`scipy.stats.mode(a)`**: Returns the mode of the array.
- **`scipy.stats.skew(a)`**: Computes the skewness of a distribution.
- **`scipy.stats.kurtosis(a)`**: Computes the kurtosis of a distribution.

```python
import numpy as np
from scipy.stats import describe

# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Descriptive statistics
stats = describe(data)
print(f"Number of elements: {stats.nobs}")
print(f"Minimum: {stats.minmax[0]}")
print(f"Maximum: {stats.minmax[1]}")
print(f"Mean: {stats.mean}")
print(f"Variance: {stats.variance}")
print(f"Skewness: {stats.skewness}")
print(f"Kurtosis: {stats.kurtosis}")
```

```python
from scipy.stats import mode

# Sample data with a mode
data = [1, 2, 2, 3, 3, 3, 4, 5, 6, 7]

# Find the mode
mode_val = mode(data)
print(f"Mode: {mode_val.mode}, Frequency: {mode_val.count}")
```

```python
from scipy.stats import skew

# Sample data
data = np.random.exponential(scale=1, size=1000)

# Compute skewness
skewness_value = skew(data)
print(f"Skewness: {skewness_value}")
```

```python
from scipy.stats import kurtosis

# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Compute kurtosis
kurtosis_value = kurtosis(data)
print(f"Kurtosis: {kurtosis_value}")
```

## 4. Statistical Tests
Statistical hypothesis tests, including:
- **`scipy.stats.ttest_1samp(a, popmean)`**: One-sample t-test.
- **`scipy.stats.ttest_ind(a, b)`**: Two-sample t-test.
- **`scipy.stats.ttest_rel(a, b)`**: Paired t-test.
- **`scipy.stats.chisquare(f_obs, f_exp)`**: Chi-squared test.
- **`scipy.stats.wilcoxon(x, y)`**: Wilcoxon signed-rank test.
- **`scipy.stats.mannwhitneyu(x, y)`**: Mann-Whitney U test.
- **`scipy.stats.kstest(rvs, cdf)`**: Kolmogorov-Smirnov test.

```python
from scipy.stats import ttest_1samp
import numpy as np

# Sample data
data = np.random.normal(0, 1, 1000)

# Perform one-sample t-test against mean=0
t_stat, p_val = ttest_1samp(data, 0)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
from scipy.stats import ttest_ind

# Sample data
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(0.5, 1, 1000)

# Perform two-sample t-test
t_stat, p_val = ttest_ind(data1, data2)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
from scipy.stats import chisquare

# Sample data: observed frequencies and expected frequencies
observed = [50, 30, 20]
expected = [40, 40, 20]

# Perform chi-squared test
chi2_stat, p_val = chisquare(observed, expected)
print(f"Chi-squared statistic: {chi2_stat}, P-value: {p_val}")
```

## 5. Correlation and Regression
Functions for calculating correlations and linear regression:
- **`scipy.stats.pearsonr(x, y)`**: Pearson correlation coefficient.
- **`scipy.stats.spearmanr(x, y)`**: Spearman rank correlation coefficient.
- **`scipy.stats.linregress(x, y)`**: Linear regression analysis.

```python
from scipy.stats import pearsonr
import numpy as np

# Sample data
x = np.random.normal(0, 1, 1000)
y = 2 * x + np.random.normal(0, 1, 1000)

# Pearson correlation
corr_coefficient, p_value = pearsonr(x, y)
print(f"Pearson correlation coefficient: {corr_coefficient}, P-value: {p_value}")
```

```python
from scipy.stats import spearmanr

# Sample data
x = np.random.normal(0, 1, 1000)
y = 2 * x + np.random.normal(0, 1, 1000)

# Spearman correlation
corr_coefficient, p_value = spearmanr(x, y)
print(f"Spearman correlation coefficient: {corr_coefficient}, P-value: {p_value}")
```

```python
from scipy.stats import linregress

# Sample data
x = np.random.normal(0, 1, 1000)
y = 3 * x + 2 + np.random.normal(0, 1, 1000)

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(f"Slope: {slope}, Intercept: {intercept}")
print(f"R-squared: {r_value**2}, P-value: {p_value}, Standard Error: {std_err}")
```

## 6. Advanced Statistical Functions
### a. F-Distribution
`scipy.stats.f`: F-distribution, used primarily in ANOVA tests.
- pdf(x, dfn, dfd): Probability density function.
- cdf(x, dfn, dfd): Cumulative distribution function.
- ppf(q, dfn, dfd): Percent point function (inverse of CDF).

```python
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the F-distribution
dfn = 3  # Numerator degrees of freedom
dfd = 5  # Denominator degrees of freedom

# Generate x values for plotting
x = np.linspace(0, 5, 100)

# Compute PDF and CDF
pdf_values = f.pdf(x, dfn, dfd)
cdf_values = f.cdf(x, dfn, dfd)

# Plot PDF and CDF
plt.figure(figsize=(12, 6))

# PDF plot
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, label='PDF', color='blue')
plt.title('F-distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')

# CDF plot
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, label='CDF', color='green')
plt.title('F-distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

### b. Multivariate Distributions
`scipy.stats.multivariate_normal`: Multivariate normal distribution.
- pdf(x, mean, cov): Probability density function for multivariate normal distribution.
- rvs(mean, cov, size): Generate random samples from a multivariate normal distribution.

```python
from scipy.stats import multivariate_normal
import numpy as np
import matplotlib.pyplot as plt

# Define mean and covariance for a 2D normal distribution
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix with correlation

# Generate grid of x, y values
x, y = np.mgrid[-3:3:.1, -3:3:.1]
pos = np.dstack((x, y))

# Compute PDF values for the grid
pdf_values = multivariate_normal.pdf(pos, mean=mean, cov=cov)

# Plot the contour of the multivariate normal PDF
plt.contour(x, y, pdf_values, cmap='viridis')
plt.title('Multivariate Normal Distribution (2D)')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Probability Density')
plt.show()
```

```python
# Generate 500 random samples from the 2D normal distribution
samples = multivariate_normal.rvs(mean=mean, cov=cov, size=500)

# Plot the generated samples
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Random Samples from Multivariate Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 7. Bayesian Statistics
`scipy.stats.beta`: Beta distribution can be used in Bayesian statistics for modeling distributions of probabilities.
`scipy.stats.gamma`: Useful for modeling waiting times and life distributions in Bayesian inference.


```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Beta distribution
a = 2  # Shape parameter alpha
b = 5  # Shape parameter beta

# Generate x values from 0 to 1
x = np.linspace(0, 1, 100)

# Compute the PDF
pdf_values = beta.pdf(x, a, b)

# Plot the Beta distribution PDF
plt.plot(x, pdf_values, label=f'Beta Distribution (a={a}, b={b})', color='blue')
plt.title('Beta Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prior Beta distribution
prior_a = 2
prior_b = 5

# Data: 3 successes, 2 failures
successes = 3
failures = 2

# Update the prior with data
posterior_a = prior_a + successes
posterior_b = prior_b + failures

# Plot the prior and posterior distributions
x = np.linspace(0, 1, 100)
prior_pdf = beta.pdf(x, prior_a, prior_b)
posterior_pdf = beta.pdf(x, posterior_a, posterior_b)

plt.plot(x, prior_pdf, label='Prior Beta Distribution', color='blue', linestyle='--')
plt.plot(x, posterior_pdf, label='Posterior Beta Distribution', color='red')
plt.title('Prior and Posterior Beta Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Gamma distribution
a = 2  # Shape parameter
scale = 2  # Scale parameter (inverse of rate)

# Generate x values
x = np.linspace(0, 15, 100)

# Compute the PDF
pdf_values = gamma.pdf(x, a, scale=scale)

# Plot the Gamma distribution PDF
plt.plot(x, pdf_values, label=f'Gamma Distribution (a={a}, scale={scale})', color='green')
plt.title('Gamma Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prior Gamma distribution
prior_a = 2
prior_scale = 2

# Data: Time until 3 events observed
observed_time = 6

# Update the prior with the data (add to the shape and scale)
posterior_a = prior_a + 3  # Add number of observed events
posterior_scale = 1 / (1 / prior_scale + observed_time)  # Update scale parameter

# Plot the prior and posterior distributions
x = np.linspace(0, 15, 100)
prior_pdf = gamma.pdf(x, prior_a, scale=prior_scale)
posterior_pdf = gamma.pdf(x, posterior_a, scale=posterior_scale)

plt.plot(x, prior_pdf, label='Prior Gamma Distribution', color='blue', linestyle='--')
plt.plot(x, posterior_pdf, label='Posterior Gamma Distribution', color='red')
plt.title('Prior and Posterior Gamma Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
