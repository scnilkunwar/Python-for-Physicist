# `scipy.fft` â€“ Fast Fourier Transforms

The `scipy.fft` module provides functions for computing Fast Fourier Transforms (FFT) and their inverse (IFFT). FFT is widely used in signal processing, image filtering, and frequency domain analysis. This tutorial covers:
- `fft` and `ifft` for one-dimensional transforms.
- `fft2` and `ifft2` for two-dimensional transforms.

### 1. One-Dimensional Fast Fourier Transforms

#### 1.1. `fft`
The `fft` function computes the one-dimensional discrete Fourier Transform.

**Syntax**:
```python
from scipy.fft import fft
transformed_signal = fft(signal)
```
- signal: Input array (time-domain signal).

````python
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt

# Create a sample signal: a sine wave
fs = 500  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector
frequency = 5 # Frequency of the sine wave
signal = np.sin(2 * np.pi * frequency * t)

# Compute the FFT
transformed_signal = fft(signal)

# Frequency vector
frequencies = np.fft.fftfreq(len(signal), d=1/fs)

# Plot the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(1, 2, 2)
plt.plot(frequencies, np.abs(transformed_signal))
plt.title('Frequency Domain Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
````

### 1.2. `ifft`
The `ifft` function computes the one-dimensional inverse discrete Fourier Transform.

**Syntax:**

```python
from scipy.fft import ifft
reconstructed_signal = ifft(transformed_signal)
```
- transformed_signal: Input array (frequency-domain signal).


````python
from scipy.fft import ifft

# Reconstruct the original signal from the FFT
reconstructed_signal = ifft(transformed_signal)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(t, reconstructed_signal.real)
plt.title('Reconstructed Signal from IFFT')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
````

## 2. Two-Dimensional Fast Fourier Transforms
### 2.1. `fft2`
The `fft2` function computes the two-dimensional discrete Fourier Transform.

**Syntax:**

```python
from scipy.fft import fft2
transformed_image = fft2(image)
```
- image: 2D input array (e.g., an image).

````python
from scipy.fft import fft2
import matplotlib.pyplot as plt
from skimage import data

# Load a sample image
image = data.camera()

# Compute the 2D FFT
transformed_image = fft2(image)

# Plot the original image and its FFT magnitude
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(np.log(np.abs(transformed_image)), cmap='gray')
plt.title('2D FFT Magnitude')
plt.axis('off')
plt.tight_layout()
plt.show()
````

### 2.2. `ifft2`
The `ifft2` function computes the two-dimensional inverse discrete Fourier Transform.

**Syntax:**

```python
from scipy.fft import ifft2
reconstructed_image = ifft2(transformed_image)
```
- transformed_image: 2D input array (frequency-domain image).

````python
from scipy.fft import ifft2

# Reconstruct the original image from the 2D FFT
reconstructed_image = ifft2(transformed_image)

# Plot the reconstructed image
plt.figure(figsize=(6, 6))
plt.imshow(reconstructed_image.real, cmap='gray')
plt.title('Reconstructed Image from IFFT')
plt.axis('off')
plt.tight_layout()
plt.show()
````
