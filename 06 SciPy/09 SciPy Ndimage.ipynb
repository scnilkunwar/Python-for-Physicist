# scipy.ndimage Module Overview

The `scipy.ndimage` module provides functions for multi-dimensional image processing and analysis.

## 1. Filtering

### Gaussian Filter
- **`scipy.ndimage.gaussian_filter(input, sigma)`**: Applies a Gaussian filter to the input array.
- **`sigma`**: Standard deviation for Gaussian kernel.

### Median Filter
- **`scipy.ndimage.median_filter(input, size)`**: Applies a median filter to the input array.
- **`size`**: Size of the filter.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, median_filter

# Create a sample image
image = np.random.rand(100, 100)

# Apply Gaussian filter
gaussian_image = gaussian_filter(image, sigma=2)

# Apply Median filter
median_image = median_filter(image, size=3)

# Plotting
plt.figure(figsize=(12, 4), dpi = 400)
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('Gaussian Filtered Image')
plt.imshow(gaussian_image, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('Median Filtered Image')
plt.imshow(median_image, cmap='gray')
plt.show()
```

## 2. Morphological Operations
### Binary Dilation
- `scipy.ndimage.binary_dilation(input, structure=None)`: Performs binary dilation on the input array.
### Binary Erosion
- `scipy.ndimage.binary_erosion(input, structure=None)`: Performs binary erosion on the input array.

```python
from scipy.ndimage import binary_dilation, binary_erosion

# Create a binary image
binary_image = np.zeros((10, 10), dtype=bool)
binary_image[4:6, 4:6] = 1

# Perform dilation and erosion
dilated_image = binary_dilation(binary_image)
eroded_image = binary_erosion(binary_image)

# Plotting
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('Original Binary Image')
plt.imshow(binary_image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('Dilated Image')
plt.imshow(dilated_image, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('Eroded Image')
plt.imshow(eroded_image, cmap='gray')
plt.show()
```

## 3. Measurements
## Labeling Connected Components
`scipy.ndimage.label(input)`: Labels the connected components in the input array.
## Measurements
`scipy.ndimage.measurements.sum(input, labels)`: Computes the sum of values in the input array for each labeled component.

```python
from scipy.ndimage import label, measurements, sum

# Create a binary image
binary_image = np.array([[0, 1, 1, 0],
                          [1, 1, 0, 0],
                          [0, 0, 1, 1]])

# Label the components
labeled_array, num_features = label(binary_image)

# Compute the sum of each component
sums = sum(binary_image, labeled_array, index=np.arange(num_features) + 1)

print(f'Labeled Array:\n{labeled_array}')
print(f'Sums of each component: {sums}')
```

## 4. Interpolation
### Zoom
`scipy.ndimage.zoom(input, zoom)`: Zooms in or out of the input array.
### Shift
`scipy.ndimage.shift(input, shift)`: Shifts the input array by the specified amount.

```python

from scipy.ndimage import zoom, shift

# Create a sample image
image = np.random.rand(10, 10)

# Zoom in
zoomed_image = zoom(image, zoom=2)

# Shift the image
shifted_image = shift(image, shift=(1, 1))

# Plotting
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('Zoomed Image')
plt.imshow(zoomed_image, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('Shifted Image')
plt.imshow(shifted_image, cmap='gray')
plt.show()
```
